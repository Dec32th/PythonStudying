## 가설 검증
### 1. 대한민국의 쓰레기 재활용 비율은 해마다 증가했을 것이다. 
위의 내용을 증명하기 위해서 4번의 표를 활용하였다. 표를 거시적으로 보여주기 위해서 4번의 내용을 이용해서 그래프로 만들었다.

#### 4. 폐기물 재활용 실적 및 업체 현황
 위의 그래프에서 년도별 재활용 폐기물량을 확인하기 위해서 표를 일부 변형했다.

```python
import pandas as pd
df = pd.read_csv('폐기물+재활용실적+및+업체현황-UTF-8.csv', encoding = 'utf-8-sig')
dff = df.loc[[1],['2010년','2011년','2012년','2013년','2014년','2015년','2016년','2017년','2018년']] 
year_list = ['2010','2011','2012','2013','2014','2015','2016','2017','2018']
Tdff = dff.transpose()
Tdff
Recycle = Tdff[1]
Recycle_list = Recycle.values.tolist()
Recycling = pd.DataFrame({"연도": year_list, "재활용량": Recycle_list})
Recycling.set_index("연도",inplace = True) 
Recycling
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
plt.title("Recyling amount")
plt.xlabel("년도")
plt.ylabel("(단위:만 톤)")
plt.plot(Recycling)
```

이를 통해서 대한민국의 재활용량은 점차 증가했음을 알 수 있다. (이때, 단위는 천 톤이다)
7번의 표를 분석해서 추가적인 결과를 얻고 싶었으나, 추가적인 결과를 얻기 위해서 분석할 표에는 사이사이에 누락된 통계가 존재하여 값이 없는 부분이 존재한다. 따라서 7번 표의 일부를 가져다 쓰고 분석하려 하였으나, 구조상 분석을 하기 힘들었다. 따라서 추가적인 국내의 자원 재활용률은 따로 추출한 아래의 표와 그래프로 대체한다.

```python
import pandas as pd
df = pd.read_csv('국내폐자원재활용률(2010~2018).csv', encoding = 'utf-8-sig')
year_list = ['2010년','2011년','2012년','2013년','2014년','2015년','2016년','2017년','2018년']
dfT =df.rename(columns={'Unnamed: 0':'품목'})
FT = dfT.drop(columns='품목')
FT.index= ['폐지', '철스크랩', '폐철캔','폐유리용기']
FT
```

```python
from matplotlib import font_manager, rc

font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

%matplotlib notebook
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(10,5))
plt.plot(FT.loc['폐지'], label = '폐지')
plt.plot(FT.loc['철스크랩'], label = '철스크랩')
plt.plot(FT.loc['폐철캔'], label = '폐철캔')
plt.plot(FT.loc['폐유리용기'], label = '폐유리용기')
plt.xlabel("년도")
plt.ylabel("(단위:%)")
plt.title("폐기물 품목별 재활용률")
plt.legend()
```

### 2. 재활용이 가능한 쓰레기 중 가장 많은 비율을 차지하는 것은 종이일 것이다.
위의 가설을 검증하기 위해서 다양한 표를 찾아보았다. 그 중에서 사용할 표는 5번의 폐자원 재활용량 활용 추이이다. 그 외의 다른 재활용품이 활용된 표를 찾아보았으나, 7번 표의 경우 데이터를 가공하기 까다롭고, 중간중간에 유실된 부분이 있어서 그 부분의 재활용량을 확인할 수 없었다. 또한, 일부 품목의 경우 자료를 하나하나 분리해야 하기 때문에, 5번 표에 존재하는 4가지 항목만 판단하기로 결정하였다.

```python
import pandas as pd
df = pd.read_csv('폐자원+재활용량+변화추이+-UTF-8.csv', encoding = 'utf-8-sig')
dff = df.loc[[3,4,5,6],['분류','2010년','2011년','2012년','2013년','2014년','2015년','2016년','2017년','2018년']]
dfT = dff.drop(columns=['분류'], axis=1)
dfT.index= ['국내철스크랩', '국내폐지', '수입폐지','국내폐유리용기']
list1 = list(dfT.loc['국내철스크랩'])
list2 = list(dfT.loc['국내폐지'])
list3 = list(dfT.loc['수입폐지'])
list4 = list(dfT.loc['국내폐유리용기'])
Fe_list = list(map(float,list1))
PK_list = list(map(float,list2))
PF_list = list(map(float,list3))
G_list = list(map(float,list4))

from matplotlib import font_manager, rc

font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

%matplotlib notebook
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(10,5))
plt.title("폐기물 종류")
plt.plot(year_list,Fe_list, label = '국내철스크랩')
plt.plot(year_list, PK_list, label ='국내폐지')
plt.plot(year_list, PF_list,label = '수입폐지')
plt.plot(year_list,G_list, label = '국내폐유리용기')
plt.xlabel("년도")
plt.ylabel("(단위:천 톤)")
plt.legend()
```

철스크랩이 재활용량이 가장 많다. 

데이터 가공 중 관련 데이터가 모두 문자열로 되어있었다. 따라서 문자열로 인식해서 그린 그래프를 1차로 다른 리스트에 값을 넣고 map함수를 이용해서 새롭게 float형으로 모두 데이터를 변환하였다. 따라서 위 같은 그래프를 새로 완성했다.

### 3. 대한민국의 폐기물 중, 사업폐기물이 제일 많을 것이다.
위의 가설을 검증하기 위해서 3번 표를 이용하였다.

```python
import pandas as pd
df = pd.read_csv('생활,+사업장(일반,+건설)+폐기물+발생+및+처리현황.csv', encoding = 'utf-8')
dff = df.loc[[1,3,4], ['2010년','2011년','2012년','2013년','2014년','2015년','2016년','2017년','2018년']]
dff.index= ['생활 폐기물', '사업장 폐기물', '건설 폐기물']
from matplotlib import font_manager, rc

font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

%matplotlib notebook
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(10,5))
plt.title("폐기물 종류별 총량")
plt.plot(dff.loc['사업장 폐기물'],label = '사업장 폐기물')
plt.plot(dff.loc['생활 폐기물'], label ='생활 폐기물')
plt.plot(dff.loc['건설 폐기물'], label = '건설 폐기물')
plt.xlabel("년도")
plt.ylabel("(단위:톤/일)")
plt.legend()
```

따라서 가설대로 사업장 폐기물이 제일 많음을 알 수 있었다. 지정폐기물과 관련된 데이터는 다른 표에 있어서 가공하기 어려웠고, 그래서 가공한 결과없이 값만을 비교했다.

## 5. 결론

1. 첫 번째 가설은 대략적으로 증가하는 추세를 띄었음을 알 수 있었으나, 그 중에서도 2014년에서 2015년은 감소하는 추세를 보였다.
또한 2015년을 기준으로 할 때, 이전보다 이후에 더 가파른 상승곡선을 띄며 증가했음을 알 수 있었다. 
#### 즉, 대한민국의 재활용 비율은 해마다 증가하였으며, 앞으로도 증가할 것이라고 예측할 수 있다. 따라서 재활용과 관련된 브랜드, 상품등도 꾸준히 증가할 것임을 예상할 수 있다.


2. 두 번째 가설은 틀렸다. 가장 많이 재활용되는 자원은 철스크랩으로, 국내외 폐지를 전부 합쳐도 철 스크랩의 양에 미치지 못했다.
#### 그러나, 가설1을 검증하고 나서 언급한 추가적인 자료에 따르면 재활용률은 폐지가 가장 높으므로, 철스크랩은 많이 사용되나, 재활용이 덜 되고 있음을 알 수 있다. 따라서 철스크랩을 재활용하기 위한 노력이 더 많이 필요할 것으로 보이고, 국가, 기업 차원에서 재활용을 위해서 더한 노력을 해야할 것이다.


3. 세 번째 가설은 정확히 맞았다. 다만, 지정 폐기물을 정확히 알 수 없어서 제대로 판단하기는 힘드나, 지정폐기물 또한 사업장 폐기물의 일환이고, 사업장 폐기물과 생활 폐기물의 단위부터 큰 차이가 남을 파악할 수 있었다.
#### 따라서 유의미한 재활용 결과를 만들고 쓰레기의 소비를 줄이기 위해서는 생활 폐기물 재활용보다는 사업장 폐기물의 재활용이 훨씬 더 중요함을 알 수 있다.


#### 결과적으로 대한민국의 재활용은 유의미한 결과를 얻고 있으며, 재활용률 또한 해마다 높아지는 편이다. 그러므로 환경 보호를 위해 우리 모두 다 재활용을 하는 것은 국가적으로 도움이 된다. 앞으로 나도 재활용이라는 과정이 귀찮더라도 조금 더 노력해서 재활용을 할 수 있도록 해야겠고, 폐기물에도 여러 종류가 있음을 자세히 알 수 있어서 좋았다.
